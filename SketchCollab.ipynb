# Task 3:Image to Pencil Sketch with Python (Using Google Colab)

## importing packages

```python
import cv2
from google.colab.patches import cv2_imshow
```

## uploading image

```python
from google.colab import files
uploaded= files.upload()
```

## reading and displaying original image

```python
image=cv2.imread("kriti.jpg")
cv2_imshow(image)
```

## converting color from BGR to RGB

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
cv2_imshow(image)
```

## converting color from RGB to grayscale

```python
img_gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
cv2_imshow(img_gray)
```

## inverting the grayscale image

Inversion here refers to focusing more on the main abject than background to enhance details and not turning the image upside down.

```python
img_invert = cv2.bitwise_not(img_gray)
cv2_imshow(img_invert)
```

## blurring inverted grayscale image

```python
img_blurred = cv2.GaussianBlur(img_invert, (21, 21), sigmaX=0, sigmaY=0)
cv2_imshow(img_blurred)
```

## creating pencil sketch

```python
pencil_sketch = cv2.divide(img_gray, img_blurred, scale=256.0)
cv2_imshow(pencil_sketch)
```
